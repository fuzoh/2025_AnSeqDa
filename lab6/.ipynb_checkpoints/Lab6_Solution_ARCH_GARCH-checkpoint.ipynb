# Solution — ARCH/GARCH Model Selection and Estimation

This notebook provides a step-by-step solution for the exercise:

1. Load `auto_1.csv` and `auto_2.csv`.
2. Test stationarity of each series (prices vs. returns).
3. Determine plausible ARCH/GARCH orders using ACF/PACF of **squared returns** and formal tests.
4. Fit competing models (ARCH(q), GARCH(p,q)), compare AIC/BIC, and interpret parameters.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.stats.diagnostic import het_arch, acorr_ljungbox

from arch import arch_model

import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (9, 3.8)

def load_series(csv_path):
    df = pd.read_csv(csv_path)
    numcols = df.select_dtypes(include=[np.number]).columns.tolist()
    if not numcols:
        raise ValueError(f"No numeric columns found in {csv_path}")
    s = df[numcols[0]].astype(float).dropna().reset_index(drop=True)
    s.name = numcols[0]
    return s, df

def maybe_to_returns(x):
    xr = x.copy()
    m, M = xr.min(), xr.max()
    mean_abs = xr.abs().mean()
    is_price = (xr.gt(0).mean() > 0.9) and (M > 5) and (mean_abs > 1.0)
    if is_price:
        r = xr.pct_change().dropna().reset_index(drop=True)
        r.name = f"return_of_{xr.name}"
        kind = 'price→return (pct_change)'
    else:
        r = xr.dropna().reset_index(drop=True)
        r.name = xr.name
        kind = 'assumed returns (no transform)'
    return r, kind

def adf_report(series, name):
    res = adfuller(series, autolag='AIC')
    stat, pval = res[0], res[1]
    print(f"ADF for {name}: stat={stat:.3f}, p-value={pval:.4f}")
    if pval < 0.05:
        print("→ Reject unit root: stationary.")
    else:
        print("→ Cannot reject unit root: non-stationary (consider differencing or use returns)." )
    return stat, pval

def summarize(series, name):
    arr = np.asarray(series, dtype=float)
    print(f"{name}: n={len(arr)}, mean={arr.mean():.6f}, std={arr.std(ddof=1):.6f}, min={arr.min():.6f}, max={arr.max():.6f}")
```

```python
auto_paths = ['auto_1.csv', 'auto_2.csv']

datasets = {}
for p in auto_paths:
    s, raw = load_series(p)
    r, how = maybe_to_returns(s)
    datasets[p] = {'raw': s, 'returns': r, 'transform': how}
    print(f"Loaded {p}: column={s.name}; transform={how}; len(raw)={len(s)}, len(returns)={len(r)}")
    summarize(r, f"returns from {p}")
    print()
```

## Stationarity tests

We test:
- ADF on the raw series and on returns (if raw looked like prices, returns should be stationary).
- Visual ACF/PACF on returns (should look close to white noise in the mean).


```python
for p, obj in datasets.items():
    print('='*70)
    print(p)
    raw = obj['raw']
    r = obj['returns']
    try:
        print('ADF on raw:')
        adf_report(raw.dropna(), f'raw[{p}]')
    except Exception as e:
        print(f'(skipped ADF on raw: {e})')
    print('ADF on returns:')
    adf_report(r, f'returns[{p}]')
    fig, ax = plt.subplots(1,2, figsize=(10,3.5))
    plot_acf(r, lags=min(30, len(r)//3), ax=ax[0])
    ax[0].set_title('ACF of returns')
    plot_pacf(r, lags=min(30, len(r)//3), ax=ax[1], method='ywm')
    ax[1].set_title('PACF of returns')
    plt.suptitle(f'Returns mean diagnostics: {p}')
    plt.show()
```

## Volatility diagnostics (squared returns)

We look for conditional heteroskedasticity:
- ACF/PACF of **squared returns**.
- Engle's ARCH LM test.
- Ljung–Box on squared returns.

Heuristics:
- PACF(squared) cutting off around `q` with ACF tailing → ARCH(q).
- Both ACF and PACF tailing slowly → GARCH(1,1) (persistence).


```python
for p, obj in datasets.items():
    print('='*70)
    print(p)
    r = obj['returns']
    rsq = r**2
    fig, ax = plt.subplots(1,2, figsize=(10,3.5))
    plot_acf(rsq, lags=min(30, len(r)//3), ax=ax[0])
    ax[0].set_title('ACF of squared returns')
    plot_pacf(rsq, lags=min(30, len(r)//3), ax=ax[1], method='ywm')
    ax[1].set_title('PACF of squared returns')
    plt.suptitle(f'Volatility diagnostics: {p}')
    plt.show()

    print('Engle ARCH LM test (up to 10 lags) on returns:')
    lm_stat, lm_pvalue, _, _ = het_arch(r, nlags=10)
    print(f'  LM stat={lm_stat:.3f}, p-value={lm_pvalue:.4g}')
    print('Ljung–Box on squared returns (lag 10):')
    lb = acorr_ljungbox(rsq, lags=[10], return_df=True)
    print(lb)
    print()
```

## Model selection by small grid search (AIC/BIC)

We fit a compact set of models and compare AIC/BIC:
- ARCH(q), q ∈ {1,2,3,4}
- GARCH(p,q), (p,q) ∈ {(1,1), (1,2), (2,1)}

We use `mean='Zero'` (returns typically have negligible mean).


```python
candidates = {
    'ARCH(1)': dict(vol='ARCH', p=1, q=0),
    'ARCH(2)': dict(vol='ARCH', p=2, q=0),
    'ARCH(3)': dict(vol='ARCH', p=3, q=0),
    'ARCH(4)': dict(vol='ARCH', p=4, q=0),
    'GARCH(1,1)': dict(vol='GARCH', p=1, q=1),
    'GARCH(1,2)': dict(vol='GARCH', p=1, q=2),
    'GARCH(2,1)': dict(vol='GARCH', p=2, q=1),
}

results = {}

for p, obj in datasets.items():
    print('='*80)
    print('Dataset:', p)
    r = obj['returns']
    r_scaled = r * 100.0
    best = None
    rows = []
    for name, spec in candidates.items():
        if spec['vol'] == 'ARCH':
            am = arch_model(r_scaled, mean='Zero', vol='ARCH', p=spec['p'])
        else:
            am = arch_model(r_scaled, mean='Zero', vol='GARCH', p=spec['p'], q=spec['q'])
        try:
            res = am.fit(disp='off')
            rows.append((name, res.aic, res.bic, res.loglikelihood))
            if (best is None) or (res.aic < best[1]):
                best = (name, res.aic, res, spec)
        except Exception as e:
            rows.append((name, np.nan, np.nan, np.nan))
            continue
    tbl = pd.DataFrame(rows, columns=['model','AIC','BIC','loglik']).sort_values('AIC')
    print(tbl.to_string(index=False))
    print(f"Selected by AIC: {best[0]}\n")
    results[p] = best
```

## Estimate selected models and interpret

We refit the selected specification and print:
- parameter estimates (ω, αᵢ, βⱼ),
- persistence (sum of α + β for GARCH),
- residual diagnostics to ensure no remaining ARCH effects,
- plot the conditional volatility.


```python
for p, (name, aic, bestres, spec) in results.items():
    print('='*80)
    print('Dataset:', p)
    r = datasets[p]['returns'] * 100.0  # scaled
    if spec['vol'] == 'ARCH':
        am = arch_model(r, mean='Zero', vol='ARCH', p=spec['p'])
    else:
        am = arch_model(r, mean='Zero', vol='GARCH', p=spec['p'], q=spec['q'])
    res = am.fit(disp='off')
    print(res.summary())
    if spec['vol'] == 'GARCH':
        params = res.params
        alpha_sum = params.filter(like='alpha').sum()
        beta_sum  = params.filter(like='beta').sum()
        print(f"Persistence sum (alpha+beta) ≈ {alpha_sum + beta_sum:.3f}")
        if alpha_sum + beta_sum < 1.0:
            print('→ Mean-reverting volatility (stationary).')
        else:
            print('→ Near-unit persistence (very slow decay).')
    z = res.std_resid.dropna()
    print('Ljung–Box on standardized residuals (lag 10):')
    print(acorr_ljungbox(z, lags=[10], return_df=True))
    print('Ljung–Box on squared standardized residuals (lag 10):')
    print(acorr_ljungbox(z**2, lags=[10], return_df=True))
    print('Engle ARCH LM on standardized residuals (up to 10 lags):')
    lm_stat, lm_pvalue, _, _ = het_arch(z, nlags=10)
    print(f'  LM stat={lm_stat:.3f}, p-value={lm_pvalue:.4g}\n')
    plt.figure(figsize=(9,3))
    res.conditional_volatility.plot()
    plt.title(f'Conditional volatility — {p} — {name}')
    plt.xlabel('Time')
    plt.ylabel('sigma_t (scaled by 100*)')
    plt.show()
```
