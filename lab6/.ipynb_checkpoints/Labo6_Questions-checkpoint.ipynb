## Section 1: The Presentation of ARCH and GARCH Models  

Classical time series models, such as AR, MA, and ARIMA, assume constant variance (or volatility) over time. However, in financial markets, volatility often changes dynamically—either increasing or decreasing based on market conditions. This variability makes traditional models less effective in capturing the true behavior of financial time series.  

To address this limitation, ARCH (Autoregressive Conditional Heteroskedasticity) and GARCH (Generalized ARCH) models were introduced. These models are specifically designed to account for time-varying volatility. As an example, let’s revisit the stock prices of Novartis and Nestle, which we analyzed in Lab 1. These datasets demonstrate how financial time series often exhibit shifts in volatility, making them ideal for applying ARCH and GARCH techniques.  


```python
# Load the stock data
data = pd.read_csv("stock_data.csv", parse_dates=['Date'])
data.head()
```

```python
# Extract the stock prices
norvatis_price = data['norvatis_price']
nestle_price = data['nestle_price']

# Plot Norvatis stock prices
plt.figure(figsize=(17, 4))
plt.plot(norvatis_price, label='Norvatis Price', color='blue')
plt.title('Stock Prices for Norvatis')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()

# Plot Nestle stock prices
plt.figure(figsize=(17, 4))
plt.plot(nestle_price, label='Nestle Price', color='green')
plt.title('Stock Prices for Nestle')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

As discussed earlier, price time series, denoted as $P_t$, are typically not stationary. To make them stationary, the first step is to apply first differencing:

$$
D_t = P_t - P_{t-1}
$$

Additionally, we define the return as:

$$
R_t = \frac{D_t}{P_{t-1}}
$$

This represents the relative change in price at time $t$. In finance, returns $R_t$ are often preferred over $D_t$ due to their clearer interpretability and relevance.  

Now, let's visualize the returns.


```python
# Calculate returns for Norvatis
norvatis_return = np.diff(norvatis_price) / norvatis_price[:-1]

# Plot returns for Norvatis
plt.figure(figsize=(17, 4))
plt.plot(norvatis_return, label='Norvatis Return', color='blue')
plt.title('Returns for Norvatis')
plt.xlabel('Time')
plt.ylabel('Return')
plt.grid(True)
plt.legend()
plt.show()

# Calculate returns for Nestle
nestle_return = np.diff(nestle_price) / nestle_price[:-1]

# Plot returns for Nestle
plt.figure(figsize=(17, 4))
plt.plot(nestle_return, label='Nestle Return', color='green')
plt.title('Returns for Nestle')
plt.xlabel('Time')
plt.ylabel('Return')
plt.grid(True)
plt.legend()
plt.show()
```

The returns are expected to be stationary by definition. Let’s verify this by performing the Augmented Dickey-Fuller (ADF) test.


```python
from statsmodels.tsa.stattools import adfuller

# Perform ADF test on Norvatis returns
adf_norvatis, pvalue_norvatis, usedlag_norvatis, nobs_norvatis, critical_values_norvatis, icbest_norvatis = adfuller(norvatis_return)
print(f"P-value for Norvatis: {pvalue_norvatis}")

# Perform ADF test on Nestle returns
adf_nestle, pvalue_nestle, usedlag_nestle, nobs_nestle, critical_values_nestle, icbest_nestle = adfuller(nestle_return)
print(f"P-value for Nestle: {pvalue_nestle}")
```

The p-values are below the significance level, confirming that the returns are stationary. This is why financial analysts often prefer working with returns instead of price values.  

However, a visual inspection of the returns reveals evidence of conditional heteroskedasticity, as we can observe volatility clustering. This phenomenon is one of the well-known “stylized facts” in financial engineering.  

In certain applications, it is important to measure and forecast these volatility patterns. This can be achieved using models that assume volatility follows an autoregressive process.



### Autoregressive Conditional Heteroskedasticity (ARCH) Models  

The ARCH model is designed to explicitly capture changes in variance over time in a time series. It models the variance as an autoregressive process, where returns are represented as white noise scaled by time-varying volatility. An ARCH(p) model is defined as:  

$$
R_{t} = Z_{t} \cdot \sigma_{t}
$$  
$$
\sigma_{t}^2 = \alpha_{0} + \alpha_{1}R_{t-1}^2 + \alpha_{2}R_{t-2}^2 + \dots + \alpha_{p}R_{t-p}^2
$$  

Here:  
- $Z_{t}$ represents innovations (or shocks) with mean 0 and variance 1, assumed to be independent of $\sigma_{t}$ for all $k \leq p$.  
- The lag length $p \geq 0$ is determined as part of the model specification, often using autocorrelation significance tests. When $p = 0$, the model reduces to a simple white noise process.  
- To ensure $\sigma_{t}^2$ remains positive, all $\alpha_{i} \geq 0$.  

The ARCH approach assumes the series is stationary, aside from changes in variance, meaning it does not contain trends or seasonal components. ARCH models are particularly useful for predicting future variance.  

---

### Generalized Autoregressive Conditional Heteroskedasticity (GARCH) Models  

The GARCH model extends ARCH by adding an autoregressive component to $\sigma_{t}^2$. This allows the model to account for dependencies not only in past returns but also in past volatility. The GARCH(p, q) model is defined as:  

$$
R_{t} = Z_{t} \cdot \sigma_{t}
$$  
$$
\sigma_{t}^2 = \alpha_{0} + \alpha_{1}R_{t-1}^2 + \alpha_{2}R_{t-2}^2 + \dots + \alpha_{p}R_{t-p}^2 + \beta_{1}\sigma_{t-1}^2 + \beta_{2}\sigma_{t-2}^2 + \dots + \beta_{q}\sigma_{t-q}^2
$$  

Key points about GARCH:  
- GARCH models generalize ARCH models, where GARCH(p, 0) reduces to an ARCH(p) model.  
- They are particularly useful for capturing long memory effects in volatility.  
- The GARCH(1,1) model, due to its simplicity and effectiveness, is one of the most widely used models in financial time series analysis.  

GARCH models are implemented in most statistical and econometric software, making them accessible and highly practical for financial volatility modeling.


## Section 2: Identifying ARCH and GARCH Models  

To build an ARCH or GARCH model, we first need to determine its structure, which involves selecting appropriate parameters (\(p\) and \(q\)). This can be done by analyzing the ACF (Autocorrelation Function) and PACF (Partial Autocorrelation Function) plots of the variance in the time series.  

Since variance is not directly observable, we use squared returns to approximate it. Squared returns are considered unbiased estimators of $\sigma_t^2$, as described by the following equation:  

$$
E(R_t^2) = E(Z_t^2 \cdot \sigma_t^2) = \sigma_t^2
$$  

By examining the ACF and PACF plots of $R_t^2$ (the squared returns), we can estimate the values of \(p\) for ARCH models and \(p, q\) for GARCH models, similar to how these plots are used for ARMA models.  

### Guidelines for Interpretation:
- If $R_t$ appears to behave like white noise and $R_t^2$ follows an AR(p) process, an ARCH(p) model may be a good choice to capture the volatility.  
- If the ACF and PACF of $R_t^2$ show a pattern resembling an ARMA structure, a GARCH model (which extends ARCH) may be more appropriate.  

However, in real-world data, these patterns are not always easy to identify, and additional diagnostic tests may be necessary to finalize the model.  

Now, let’s use the returns from our data to determine the order \(p\) and \(q\) for an ARCH or GARCH model.


### - For Novartis Returns  

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import scipy.stats as stats
#import pylab

# Create a side-by-side layout for the time series and Q-Q plot
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Time series plot of Novartis returns
axes[0].plot(norvatis_return, color='tab:blue')
axes[0].set_title('Novartis Returns Over Time')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Return')
axes[0].grid(True)

# Q-Q plot for normality check
stats.probplot(norvatis_return, dist="norm", plot=axes[1])
axes[1].set_title('Q-Q Plot for Novartis Returns')

plt.tight_layout()
plt.show()


# ACF and PACF plots side by side
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# ACF plot
plot_acf(norvatis_return, lags=30, ax=axes[0])
axes[0].set_title('ACF for Novartis Returns')
axes[0].grid(True)

# PACF plot
plot_pacf(norvatis_return, lags=30, ax=axes[1])
axes[1].set_title('PACF for Novartis Returns')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

**Observation.** The ACF and PACF plots above suggest that the returns resemble white noise. This means there are no significant autocorrelations, except at lag 0, which is expected for a stationary series with no clear patterns over time.  

Next, we’ll analyze the squared returns by plotting their ACF and PACF. This will help us identify any potential patterns in the variance, which are important for selecting models like ARCH or GARCH.


```python
# Side-by-side plots: Squared returns and Q-Q plot
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Plot squared returns
axes[0].plot(norvatis_return**2, color='purple')
axes[0].set_title('Squared Returns for Novartis')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Squared Return')
axes[0].grid(True)

# Q-Q plot for squared returns
stats.probplot(norvatis_return**2, dist="norm", plot=axes[1])
axes[1].set_title('Q-Q Plot for Squared Novartis Returns')

plt.tight_layout()
plt.show()

# Side-by-side plots: ACF and PACF for squared returns
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# ACF plot for squared returns
plot_acf(norvatis_return**2, lags=30, ax=axes[0])
axes[0].set_title('ACF for Squared Novartis Returns')
axes[0].grid(True)

# PACF plot for squared returns
plot_pacf(norvatis_return**2, lags=30, ax=axes[1])
axes[1].set_title('PACF for Squared Novartis Returns')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

**Observation.** The ACF and PACF plots of the squared series indicate a pattern resembling an ARMA process. This is because neither plot shows a complete cutoff after a specific number of lags. Such a pattern suggests that a **GARCH(1,1)** model could be a suitable choice for capturing the dynamics of the variance.


### - For Nestle Returns  

```python
# Side-by-side plots: Returns and Q-Q plot for Nestle
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Plot Nestle returns
axes[0].plot(nestle_return, color='tab:blue')
axes[0].set_title('Returns for Nestle')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Return')
axes[0].grid(True)

# Q-Q plot for Nestle returns
stats.probplot(nestle_return, dist="norm", plot=axes[1])
axes[1].set_title('Q-Q Plot for Nestle Returns')

plt.tight_layout()
plt.show()

# Side-by-side plots: ACF and PACF for Nestle returns
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# ACF plot for Nestle returns
plot_acf(nestle_return, lags=30, ax=axes[0])
axes[0].set_title('ACF for Nestle Returns')
axes[0].grid(True)

# PACF plot for Nestle returns
plot_pacf(nestle_return, lags=30, ax=axes[1])
axes[1].set_title('PACF for Nestle Returns')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

**Observation.** The ACF and PACF plots for Nestle returns suggest that the series behaves like white noise, with no significant autocorrelations beyond lag 0.  

Now, let’s analyze the squared returns by plotting their correlogram. This will help us explore potential patterns in the variance.


```python
# Side-by-side plots: Squared returns and Q-Q plot for Nestle
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Plot squared returns for Nestle
axes[0].plot(nestle_return**2, color='green')
axes[0].set_title('Squared Returns for Nestle')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Squared Return')
axes[0].grid(True)

# Q-Q plot for squared returns
stats.probplot(nestle_return**2, dist="norm", plot=axes[1])
axes[1].set_title('Q-Q Plot for Squared Nestle Returns')

plt.tight_layout()
plt.show()

# Side-by-side plots: ACF and PACF for squared returns
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# ACF plot for squared returns
plot_acf(nestle_return**2, lags=30, ax=axes[0])
axes[0].set_title('ACF for Squared Nestle Returns')
axes[0].grid(True)

# PACF plot for squared returns
plot_pacf(nestle_return**2, lags=30, ax=axes[1])
axes[1].set_title('PACF for Squared Nestle Returns')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

**Observation.** For Nestle, the exact values of \( p \) and \( q \) are not immediately clear from the plots. However, it is evident that the model does not resemble a simple AR(p) process. Based on the observed patterns, we propose using a **GARCH(1,1)** model.

## Section 3: Estimating GARCH and ARCH Models in Python  

To define a GARCH or ARCH model, we use the `arch_model()` function. This function allows us to specify:  
1. The model for the **mean** of the series. For our case, setting `mean='Zero'` is appropriate, as we assume the mean of the returns is zero.  
2. The model for the **variance** of the series. We can choose either `vol='ARCH'` for an ARCH model or `vol='GARCH'` for a GARCH model.  
3. The **lag parameters**, which determine the order of the ARCH or GARCH model (e.g., GARCH(1,1)).  

Now, let’s estimate a GARCH(1,1) model for our series.


```python
# Ensure the `arch` library is installed: pip install arch
from arch import arch_model

# Define a GARCH(1,1) model for Novartis returns
garch_model = arch_model(norvatis_return, vol='GARCH', p=1, q=1, mean='Zero')

# Fit the model to the data
garch_result = garch_model.fit()

# Print the summary of the fitted model
print(garch_result.summary())
```

**Observation.** From the table, the estimated coefficients are as follows:  
- $\alpha_{0} = 9.9483 \times 10^{-6}$  
- $\alpha_{1} = 0.10$  
- $\beta_{1} = 0.80$  

Next, we can visualize the residuals and the fitted volatilities to better understand the model’s performance.


```python
# Plot the results of the fitted GARCH model with daily annualization and adjust the figure size
fig = garch_result.plot(annualize="D")
```

```python
# Define a GARCH(1,1) model for Nestle returns
garch_model_nestle = arch_model(nestle_return, vol='GARCH', p=1, q=1, mean='Zero')

# Fit the model to the Nestle data
garch_result_nestle = garch_model_nestle.fit()

# Print the summary of the fitted model
print(garch_result_nestle.summary())
```

```python
# Visualize the residuals and the fitted volatilities for the Nestle model
fig = garch_result_nestle.plot(annualize="D")
```

**Observation:** The GARCH(1,1) model appears to be a good choice for capturing the time-varying variance in both time series. However, a detailed analysis of the residuals is necessary to confirm the model's adequacy.
