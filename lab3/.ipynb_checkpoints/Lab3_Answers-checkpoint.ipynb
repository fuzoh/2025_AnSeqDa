```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the datasets
auto_1 = pd.read_csv("auto_1.csv", names=['value'])
auto_2 = pd.read_csv("auto_2.csv", names=['value'])

# Create a time variable 'mytime' for each dataset based on their individual lengths
auto_1['mytime'] = np.arange(len(auto_1))
auto_2['mytime'] = np.arange(len(auto_2))
```

```python
auto_1.head()
```

```python
auto_2.head()
```

```python
print('auto_1.shape = ', auto_1.shape)
print('auto_2.shape = ', auto_2.shape)
```

```python
# Visualize the time series for both datasets
plt.figure(figsize=(12, 6))

# Plot for auto_1
plt.subplot(2, 1, 1)
plt.plot(auto_1['mytime'], auto_1['value'], label='auto_1')
plt.title('Time Series for auto_1')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid()

# Plot for auto_2
plt.subplot(2, 1, 2)
plt.plot(auto_2['mytime'], auto_2['value'], label='auto_2', color='tab:orange')
plt.title('Time Series for auto_2')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Stationarity test for auto_1
adf_result_1 = adfuller(auto_1['value'])
pvalue_1 = adf_result_1[1]
print(f"ADF test p-value for auto_1: {pvalue_1:.4f}")

# Plot the correlogram (ACF and PACF) for auto_1
fig, axes = plt.subplots(1, 2, figsize=(17, 5))

# ACF for auto_1
plot_acf(auto_1['value'], lags=14, alpha=0.05, ax=axes[0])
axes[0].set_title('Autocorrelation Function (ACF) for auto_1')

# PACF for auto_1
plot_pacf(auto_1['value'], lags=14, alpha=0.05, ax=axes[1])
axes[1].set_title('Partial Autocorrelation Function (PACF) for auto_1')

plt.tight_layout()
plt.show()
```

**Observation.** For **auto_1**, the optimal order for **p** is clearly 2, based on the partial autocorrelation function (PACF) plot.


```python
# Stationarity test for auto_2
adf_result_2 = adfuller(auto_2['value'])
pvalue_2 = adf_result_2[1]
print(f"ADF test p-value for auto_2: {pvalue_2:.4f}")

# Plot the correlogram (ACF and PACF) for auto_2
fig, axes = plt.subplots(1, 2, figsize=(17, 5))

# ACF for auto_2
plot_acf(auto_2['value'], lags=14, alpha=0.05, ax=axes[0])
axes[0].set_title('Autocorrelation Function (ACF) for auto_2')

# PACF for auto_2
plot_pacf(auto_2['value'], lags=14, alpha=0.05, ax=axes[1])
axes[1].set_title('Partial Autocorrelation Function (PACF) for auto_2')

plt.tight_layout()
plt.show()

```

**Observation.** For **auto_2**, the optimal order for **q** appears to be 1, as suggested by the autocorrelation function (ACF) plot.


```python
import statsmodels.api as sm

# Fit the AR(2) model using Maximum Likelihood Estimation (MLE) on auto_1
# Using ARIMA with order (p=2, d=0, q=0) to represent an AR(2) model with no differencing or MA component
model_auto_1 = sm.tsa.ARIMA(auto_1['value'], order=(2, 0, 0), trend='n').fit()

# Display the model summary
print(model_auto_1.summary())
```

```python
# Fit the MA(1) model using Maximum Likelihood Estimation (MLE) on auto_2
# Using ARIMA with order (p=0, d=0, q=1) to represent an MA(1) model with no differencing or AR component
model_auto_2 = sm.tsa.ARIMA(auto_2['value'], order=(0, 0, 1), trend='n').fit()

# Display the model summary
print(model_auto_2.summary())
```
